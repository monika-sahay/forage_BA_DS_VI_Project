# Exploratory Data Analysis on Customer Bookings data for British Airways

```python
#imports

import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get current working directory

cwd = os.getcwd()

#read the csv

df = pd.read_csv(cwd + "/customer_booking.csv",  encoding="ISO-8859-1")
```

## Sales Channel

```python
per_internet = df.sales_channel.value_counts().values[0]  / df.sales_channel.count() *100
per_mobile = df.sales_channel.value_counts().values[1]  / df.sales_channel.count() *100
```

```python
# Create the pie chart
plt.figure(figsize=(4, 4))  # Set the figure size
plt.pie([per_internet,per_mobile], labels=['Internet','Mobile'], autopct='%1.1f%%', startangle=90)  # Create the pie chart
plt.axis('equal')  # Ensure the pie chart is circular

# Add a title
plt.title('Distribution of sales')

# Show the chart
plt.show()
```

## Trip Type

```python
per_round = df.trip_type.value_counts().values[0]/ df.trip_type.count() *100
per_oneway = df.trip_type.value_counts().values[1]/ df.trip_type.count() *100
per_circle = df.trip_type.value_counts().values[2]/ df.trip_type.count() *100
```

```python
print(f"Percentage of round trips: {per_round} %")
print(f"Percentage of One way trips: {per_oneway} %")
print(f"Percentage of circle trips: {per_circle} %")
```

## Purchase Lead

```python
plt.figure(figsize=(15,5))
sns.histplot(data=df, x="purchase_lead", binwidth=20,kde=True)
```

It is uncommon to make bookings more than 2 years in advance of the travel date. However, this could be attributed to cancellations and rebookings within a 6-month period. Normally, airlines only allow ticket rebooking within a year. Nevertheless, we will treat these cases as outliers, as they can significantly impact the results of our predictive model.

```python
(df.purchase_lead >600).value_counts()
```

To ensure that customers are not booking more than 1 and a half years in advance, we will exclude any entries with a purchase_lead time exceeding 600 days.

```python
df[df.purchase_lead > 600]
```

```python
#filtering the data to have only purchase lead days less than 600 days
df = df[df.purchase_lead <600 ]
```

## Length Of Stay

```python
plt.figure(figsize=(15,5))
sns.histplot(data=df, x="length_of_stay", binwidth=15,kde=True)
```


Let's examine the number of entries that have a length of stay exceeding 100 days.

```python
(df.length_of_stay> 200).value_counts()
```

```python
df[df.length_of_stay> 500].booking_complete.value_counts()
```

To make a decision regarding whether to remove entries with a length of stay exceeding 600 days, we need to consider additional business knowledge. There could be various reasons for such bookings, and it requires further analysis. However, for our current focus, we will only consider bookings with a length of stay less than 500 days.

```python
#filtering the data to have only length of stay days less than 500 days
df = df[df.purchase_lead <500 ]
```

## Flight Day
We will map the flight day with a number of a week.

```python
mapping = {
    "Mon" : 1,
    "Tue" : 2,
    "Wed" : 3,
    "Thu" : 4,
    "Fri" : 5,
    "Sat" : 6,
    "Sun" : 7
}

df.flight_day = df.flight_day.map(mapping)
```

```python

df.flight_day.value_counts()
```

Most of the customers want to travel on Monday and choose Saturday as least preffered day as flight day.

## Booking Origin

```python
plt.figure(figsize=(15,5))
ax = df.booking_origin.value_counts()[:20].plot(kind="bar")
ax.set_xlabel("Countries")
ax.set_ylabel("Number of bookings")
```

The chart illustrates the country from which travelers had the highest number of booking applications.

```python
plt.figure(figsize=(15,5))
ax = df[df.booking_complete ==1].booking_origin.value_counts()[:20].plot(kind="bar")
ax.set_xlabel("Countries")
ax.set_ylabel("Number of complete bookings")
```

The chart displays the country from which travelers had their bookings completed.

## Booking complete

```python
successful_booking_per = df.booking_complete.value_counts().values[0] / len(df) * 100
```

```python
unsuccessful_booking_per = 100-successful_booking_per
```

```python
# Create the pie chart
plt.figure(figsize=(4, 4))  # Set the figure size
plt.pie([successful_booking_per,unsuccessful_booking_per], labels=['Sucessful bookings','unsuccessful bookings'], autopct='%1.1f%%', startangle=90)  # Create the pie chart
plt.axis('equal')  # Ensure the pie chart is circular

# Add a title
plt.title('Distribution of sales')

# Show the chart
plt.show()
print(f"Out of 50000 booking entries only {round(unsuccessful_booking_per,2)} % bookings were successfull or complete.")
```
